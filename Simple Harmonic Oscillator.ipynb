## Numeric Intergration of Harmonic Oscillator

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

```python
t = np.linspace(1,15,1000) 
omega_sq = 1
y = [0,2] # initial condition y[0]=x, y[1]=v
```

```python
def harmonic(t,y):
    return [y[1],-omega_sq*y[0]]

sho = solve_ivp(harmonic, [0,1000], y0 = y, t_eval = t)
```

```python
plt.plot(t,sho.y[0])
plt.ylabel("Position[m]")
plt.xlabel("Time [s]")
plt.title('Simple Harmonic Oscillator', fontsize = 12)
plt.grid()
```

```python
plt.plot(t,sho.y[1])
plt.ylabel("Velocity [m/s]")
plt.xlabel("Time[s]")
plt.title("Simple Harmonic Oscillator", fontsize = 12)
plt.grid()
```
