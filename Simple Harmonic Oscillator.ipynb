## Numeric Intergration of Harmonic Oscillator

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

#### Defining some constants and initial conditions

```python
t = np.linspace(0,16,100) # range of integration
m = 2 # mass
k = 10 # spring konstant
omega_sq = k/m # angular frequenzy squared
g = 9.81 # gravitational accerlaration
y = [m*g/k,2] # initial condition y[0]=x, y[1]=v
```

#### Defining the function to be integrated as a vector and integrate it using scipy _ivp

```python
def harmonic(t,y):
    return [y[1],-omega_sq*y[0] + m*g]

solution = solve_ivp(harmonic, (0,16), y0 = y, t_eval = t)
```

#### Plotting the Position

```python
plt.plot(t,solution.y[0])
plt.ylabel("Position[m]")
plt.xlabel("Time [s]")
plt.title('Simple Harmonic Oscillator', fontsize = 12)
plt.grid()
```

#### Plotting the Velocity

```python
plt.plot(t,solution.y[1])
plt.ylabel("Velocity [m/s]")
plt.xlabel("Time[s]")
plt.title("Simple Harmonic Oscillator", fontsize = 12)
plt.grid()
```

### Showing energy conservation

```python
# E = 1/2mv^2 +1/2kx^2 - mgx = const.

E_0 = 1/2*m*y[1]**2 + 0.5*k*(y[0])**2 - m*g*y[0] # energy a t=0
print(f"The energy at t=0 is {E_0}")
E_t = 0.5*m*solution.y[1]**2 + 0.5*k*solution.y[0]**2 - m*g*solution.y[0] # energy at time t
plt.plot(t,E_t)
```
