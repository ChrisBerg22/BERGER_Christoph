#### We want to visualize the solution of the Schrödinger-equation for the quantum mechanical infinte square well 

The solution of the time-independent Schrödinger-equation $\left(-\frac{\hbar}{2*m}\nabla^2 + V(x)\right)\psi(x) = E\psi(x)$ is given by: $\psi(x) = \sqrt{\frac{2}{a}}\cdot \sin(\frac{n\pi x}{a}) $ , where a is the lenght of the square well and n the excitation value.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# We want to plot the first first wave function of this system
plt.style.use(['science','notebook', 'grid'])

def psi(x,a,n):
    '''wave function of the problem '''
    return (np.sqrt(2/a))*np.sin(n*np.pi*x/a)
def plot(x,a,n):
    ''' plof of the wave function '''
    plt.plot(x, psi(x,a,n), label = f'n = {n}')
    plt.legend(loc = 'lower left')
    plt.title('Wave function $\psi(x)$')
    plt.xlabel('x')
    plt.ylabel('$\psi(x)$')
def main():
    '''excecutes the program '''
    plt.figure(figsize=(10,5))
    a = 4
    x = np.linspace(0,4,100)
    for i in range(1,4):
        plot(x,a,i)
```

```python
main()
```
