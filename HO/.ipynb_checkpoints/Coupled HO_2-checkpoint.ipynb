### Numeric Integration of the coupled harmonic Oscillator

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import solve_ivp
```

#### Defining some constants and initial condition

```python
# mass 1 and 2 are set to be 1
g = 9.81
m = 2
x1_0 = m*g/k1 # position of mass 1
x2_0 = m*g/k2 # -''- of mass 2
v1_0 = 2 # velocity of mass 1
v2_0 = 4 # -''- of mass 2
S_0 = [x1_0,x2_0,v1_0,v2_0]
k1 = 50 # spring konstants
k2 =60
```

### Defining the system of coupled equations in vector form

```python
def dSdt(t,S):
    x1, x2, v1, v2 = S
    return [v1,v2,-(k1 + k2)*x1/m + k2*x2/m + m*g,k2*x1/m - k2*x2/m + m*g]
```

### Solving using scipy solve_ivp

```python
t = np.linspace(0,5,100) # range of integration

sol = solve_ivp(dSdt,t_span=(0,max(t)), y0=S_0,t_eval=t) # solution
```

```python
x1 = sol.y[0]
x2 = sol.y[1]
v1 = sol.y[2]
v2 = sol.y[3]

plt.plot(t,x1,'k', label = "mass 1")
plt.plot(t,x2, 'r', label = "mass 2")

plt.title("Coupled HO")
plt.xlabel("Time t [s]")
plt.ylabel("Position")
plt.grid()
plt.legend()
```

### Energy conservation

```python
E_t = 0.5*(v1**2 + v2**2) + 0.5*k1*x1**2 + 0.5*k2*(x1 - x2)**2 + 0.5*k2*(x2 - x1)**2 + m*g*x1 + m*g*x2 # total energy 
E_kin = 0.5*(v1**2 + v2**2) # kinetic energy
E_pot = 0.5*k1*x1**2 + 0.5*k2*(x1 - x2)**2 + 0.5*k2*(x2 - x1)**2 + m*g*x1 + m*g*x2 # potential energy

plt.plot(t,E_t, label = 'total E')
plt.plot(t,E_kin, 'k', label = 'kinetic E')
plt.plot(t,E_pot,'r', label = 'potential E')

plt.title("Coupled HO")
plt.xlabel("Time t [s]")
plt.ylabel("Energy E [J]")
plt.grid()
plt.legend()

E_mean = np.mean(E_t)
E_0 = 0.5*(v1_0**2 + v2_0**2) + 0.5*k1*x1_0**2 + 0.5*k2*(x1_0 - x2_0)**2 + 0.5*k2*(x2_0 - x1_0)**2 + m*g*x1_0 + m*g*x2_0

print(f"The energy at t=0 is {E_0}J")
print(f"The mean energy at an arbitrary t is {E_mean}J")

# The kinetic energy is much smaller than the potential energy and the total energy. We see that the potential energy and the 
# total energy are pretty much the same. Therefore we can argue that the energy is conserved. Also we see, that at some points 
# the potential energy has a maximum, as the kintetic energy has its minimum, which makes sense. 
```
