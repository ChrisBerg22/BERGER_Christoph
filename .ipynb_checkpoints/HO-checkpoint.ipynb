## Harmonic Motion without Gravity

```python
import numpy as np
from scipy.integrate import solve_ivp
import matplotlib.pyplot as plt
```

```python
t = np.linspace(0,16,100) # range of integration
m = 2 # mass
k = 10 # spring konstant
omega_sq = k/m # angular frequenzy squared
y = [0,2] # initial condition y[0]=x, y[1]=v
```

```python
def harmonic(t,y):
    return [y[1],-omega_sq*y[0]]

solution = solve_ivp(harmonic, (0,16), y0 = y, t_eval = t)
```

```python
plt.plot(t,solution.y[0])
plt.ylabel("Position[m]")
plt.xlabel("Time [s]")
plt.title('Simple Harmonic Oscillator', fontsize = 12)
plt.grid()
```

```python
plt.plot(t,solution.y[1])
plt.ylabel("Velocity [m/s]")
plt.xlabel("Time[s]")
plt.title("Simple Harmonic Oscillator", fontsize = 12)
plt.grid()
```

```python
# E = 1/2mv^2 +1/2kx^2 = const.

E_0 = 0.5*m*y[1]**2 
print(f" The energy of the simple Harmonic Oscillator is {E_0}J")
E_t = 0.5*m*solution.y[1]**2 + 0.5*k*solution.y[0]**2
plt.plot(t,E_t)
plt.xlabel("Time t [s]")
plt.ylabel("Energy E [J]")
plt.grid()
plt.title("Harmonic motion", fontsize = 12)
```
